## 1 Этат теста

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame(np.array([[0, 2, 14, 10 ,14],
                             [0, 2, 8, 14 ,16],
                             [4, 6, 9, 11 ,10],
                             [12, 15, 4, 5 ,4],
                             [2, 10, 11, 9 ,8],
                             [6, 8, 14, 5 ,7],
                             [7, 5, 9, 9 ,10],
                             [10, 9, 9, 5 ,7],
                             [0, 6, 10, 12 ,12],
                             [7, 8, 11, 4 ,10]]),columns=[1, 2, 3, 4, 5],
                              index=['Шпагаты',
                                     'Поочередные махи ногами',
                                     'Базовые движения рук',
                                     'Блок обязательных «чир прыжков»',
                                     'Блок обязательных «лип прыжков»',
                                     'Разнообразные пируэты различной сложности',
                                     'Акробатические элементы',
                                     'Контроль точности и координации движений',
                                     'Контроль чувства ритма',
                                     'Контроль за командной синхронностью'])
```

```python
df1['X'] = (df1[1]*1 +df1[2]*2 + df1[3]*3+ df1[4]*4+df1[5]*5)/(df1[1] +df1[2] + df1[3]+ df1[4]+df1[5])
```

```python
df1
```

#### Вычисление дисперсии оценок студентов D

Сумма квадратов отклонений деленное на число студентов

```python
df1['D'] = (((df1['X']-1)**2)*df1[1] + ((df1['X']-2)**2)*df1[2] + ((df1['X']-3)**2)*df1[3] + ((df1['X']-4)**2)*df1[4] + ((df1['X']-5)**2)*df1[5])/40
```

```python
df1
```

#### Вычисления стандартного отклонения оценок студентов σ (это конень от D)


```python
df1['σ'] = df1['D']**0.5
```

```python
df1
```

#### Вычисления m 

```python
df1['m'] = df1['σ']/40**0.5
```

```python
df1
```
